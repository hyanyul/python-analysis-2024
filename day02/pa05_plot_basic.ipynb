````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
## 단일차트
plt.plot([1, 2, 3, 4, 5])
plt.show()  # [<matplotlib.lines.Line2D at 0x2082e7b8d70>] 와 같은 객체 정보 출력이 사라짐
````

````python
plt.plot(np.random.randn(50))   # np.random.randn(n) = n개의 정수를 랜덤하게 만들어 줌
plt.show()
````

````python
## 다중차트, figure() 사용, 4개 이상 사용 불가
fig = plt.figure()
plt1 = fig.add_subplot(2, 2, 1) #1번 차트
plt2 = fig.add_subplot(2, 2, 2) #2번 차트
plt3 = fig.add_subplot(2, 2, 3) #3번 차트
plt4 = fig.add_subplot(2, 2, 4) #4번 차트

plt1.plot(np.random.randn(50))
plt2.hist(np.random.randn(100))
plt3.scatter(np.arange(30), np.random.randn(30))
plt4.plot(np.random.randn(30), 'violet')    #색상 'r', 'g', 'b', 'y', 'violet'

plt.show()
````

````python
#다중차트 2번째, 4개 이상 가능
fig, axes = plt.subplots(2, 3, figsize = (18, 6))   #subplots() figsize로 차트 넓이 조정 가능
axes[0][0].plot(np.random.randn(50))
axes[0][1].hist(np.random.randn(100))
axes[1][0].scatter(np.arange(30), np.random.randn(30))
axes[1][1].plot(np.random.randn(30), 'pink')

plt.show()
````

#### 차트 스타일 지정 옵션
- label, style, alpha, rot, xticks, ylticks, xlim, ylim, color, marker 등

````python
fig, axes = plt.subplots(2, 3, figsize = (18, 6))

#color='k': 검정색, 'crimson': 핏빛 / alpha: 투명도(0.0~1.0)
axes[0][0].plot(np.random.randn(50), color='k', alpha=0.1)  
axes[0][1].hist(np.random.randn(100), color='g', bins=30)   #bins: 그루핑 개수
axes[0][2].plot(np.random.randn(50), color='b', linestyle='dashed') #linestyle: 선 종류(-, --, -. 등 / solid, dotted, dashed 등)

#marker *(별), ^(삼각형), o(동그라미)
axes[1][0].scatter(np.arange(30), np.random.randn(30), marker='*')
axes[1][1].plot(np.random.randn(30), 'pink')
axes[1][2].plot(np.random.randn(30), 'b', marker='^')

plt.show()
````

````python
#cumsum() cumulative sum(누적합)
#플롯에 데이터 여러 개 표시되면 파란색, 주황색, 녹색 순
plt.plot(np.random.randn(50).cumsum(), label='data1')
plt.plot(np.random.randn(50).cumsum(), label='data2')
plt.plot(np.random.randn(50), label='data3')

#legend() 범례 표시 함수
#위치값: 'best', 'upper right', 'upper left', 'lower lift', 'lower right', 'right', 'center left',
#'center right', lower center' ,'upper center', 'center'
plt.legend(loc='best')
plt.show()
````

````python
#x축 설정
plt.figure(figsize=(18, 5))
plt.plot(np.random.randn(1000))
plt.xticks([0, 250, 500, 750, 1000], labels=['min', '25%', 'median', ' 75%', 'max'])

#y축 설정도 동일    xticks -> yticks

plt.title('Statics')  #한글은 기본적으로 지원안돼서 깨짐

plt.show()
````

````python
#내장 데이터셋 목록 확인
sns.get_dataset_names()
````

````python
#seaborn 스타일 지정
sns.set_style("whitegrid")   #"white", "dark", "whitegrid", "darkgrid", "ticks"
````

````python
#seaborn 컬러 코드
sns.palplot(sns.color_palette("tab10"))
# sns.set_color_codes('tab10')
````

````python
# irisData = sns.load_dataset('iris')    #붓꽃 샘플 데이터
flights = sns.load_dataset('flights')   #1949~1960년 월별 비행기 승객 수
````

````python
# irisData.plot()
flights[['month', 'passengers']].plot()
````

````python
sns.barplot(data=flights, x='year', y='passengers', color='r')
plt.show()
````

````python
sns.boxplot(data=flights, x='year', y='passengers')
plt.show()
````

````python
sns.swarmplot(data=flights, x='year', y='passengers')
plt.show()
````

````python
sns.lineplot(data=flights, x='year', y='passengers')
plt.show()
````

````python
sns.distplot(flights['passengers']) # distplot -> seaborn v0.14.0.에서 삭제 예정(현재 버전: 0.13.2)
plt.show()
````
